```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
car_data=pd.read_csv("data/car data.csv")
print(car_data.head())
```

```python
print(car_data.shape)
```

```python
print(car_data.info())
```

```python
print(car_data.isnull().sum())
```

```python
print(car_data.Fuel_Type.value_counts())
```

```python
print(car_data.Seller_Type.value_counts())
```

```python
print(car_data.Transmission.value_counts())
```

```python
car_data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
car_data.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
car_data.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
print(car_data.head())
```

```python
X=car_data.drop(['Car_Name','Selling_Price'],axis=1)
Y=car_data['Selling_Price']
```

```python
# print(X)
# print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
len_model=LinearRegression()
```

```python
len_model.fit(X_train,Y_train)
```

```python
train_data_prediction = len_model.predict(X_train)
train_error_score=metrics.r2_score(Y_train,train_data_prediction)
```

```python
test_data_prediction=len_model.predict(X_test)
test_error_score=metrics.r2_score(Y_test,test_data_prediction)
```

```python
print(train_error_score)
```

```python
print(test_error_score)
```

```python
plt.bar(Y_train,train_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Prediction Price")
plt.title("Actual Car price Vs Prediction Car Price")
plt.show()
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Prediction Price")
plt.title("Actual Car price Vs Prediction Car Price")
plt.show()
```
